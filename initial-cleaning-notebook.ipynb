```python
p1 = pd.get_dummies(df['party_1'])
p2 = pd.get_dummies(df['party_2'])
p3 = pd.get_dummies(df['party_3'])
p4 = pd.get_dummies(df['party_4'])
promotions = pd.get_dummies(df['promotion'])
take_bigger = lambda s1, s2: s1 if s1.sum() > s2.sum() else s2
```

```python
dummies_1 = p1.combine(p2, take_bigger, fill_value= 0, overwrite = False)
dummies_2 = p3.combine(p4, take_bigger, fill_value = 0, overwrite = False)
dummies_fin = dummies_1.combine(dummies_2, take_bigger, fill_value = 0, overwrite = False)
```

```python
dummies_fin.drop(columns = '(?)', inplace = True)
```

```python
#removes all points were the wide participant net caught some accidental noise
for column in dummies_fin.columns:
    if column in ['World Wrestling Entertainment', 'Total Nonstop Action Wrestling', 
                  'CHIKARA', 'NWA World Championship Wrestling', 'Comments',
                 'Continental Wrestling Association', 'Extreme Championship Wrestling',
                 '%%HTML', 'Lucha Underground', 'National Wrestling Alliance Total Nonstop Action',
                 'Ohio Valley Wrestling', 'README.md', '%%SVG', 'Untitled.ipynb','World Championship Wrestling',
                 'World Wide Wrestling Federation', 'World Wrestling Federation', 'Wrestling Society X',
                 ]:
        dummies_fin.drop(columns = column, inplace = True)
    
```

```python
df = pd.concat([df,dummies_fin, promotions], axis = 1).drop(columns = ['promotion','party_1','party_2','party_3', 'party_4'])
```

```python
df.head()
```

```python
for column in df.columns:
  if '.' in column:
    #df.withColumnRenamed([column], df[column](lambda x: x.replace(u'.','')))
    df.rename(mapper = lambda x: x.replace(u'.',''), axis = 1, inplace = True)
```
