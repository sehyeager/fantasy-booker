```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/dummied_data.csv')
df.head()
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])
```

```python
df['duration'] = df['end_date'] - df['start_date']
```

```python
df.describe()
```

```python
plt.title('distribution of ratings')
plt.hist(df['rating']);
#interesting to note that the ratings peak around 6.5-7
```

```python
plt.figure(figsize= (15,15))
plt.title('distribution of time in nanoseconds')
plt.hist(df['duration'].astype(int), bins = 'auto');
#the peak is about 10 quadrillion nanoseconds, or approximately 115 days.  
```

```python
plt.figure(figsize = (40,40))
sns.heatmap(df.drop(columns = ['title','start_date','end_date', 'duration']), cmap = 'flag');
#the spike in the middle of the 's's and the end of the promotions is likely an aggregate of everyone these
```
